üêº Challenge 1. Warm up
We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

From 0 to 999 ==> 1
From 1000 to 9999 ==> 2
From 10000 to 99999 ==> 3
From 100000 to 999999 ==> 4
More than 1000000 ==> 5

```python
import pandas as pd
import hashlib
```

```python
df1 = pd.read_csv('./data/input/IMDB-Movie-Data.csv')
```

```python
df1
```

```python
def bins(x):
    if 0 < x < 999:
        return 1
    elif 1000 < x < 9999:
        return 2
    elif 10000 < x < 99999:
        return 3
    elif 100000 < x < 999999:
        return 4
    elif x > 1000000:
        return 5
```

```python
df1['bins'] = df1['Votes'].apply(bins)
```

üêº üêº üêº Challenge 3. Using the lambda
We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

```python
def new_ranking(x,y):
    if 'Thriller' in x:
        return y+1
    elif 'Comedy' in x:
        return y-1
    else:
        return y
```

üêº üêº üêº üêº Challenge 4. Now the real stuff
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.



```python
def ascii_function(x,y):
    y = int(y)
    sum_ascii_values = 0
    for i in x:
        sum_ascii_values += ord(i)
    div = int(sum_ascii_values/y)
    if div == 1 or div == 0 or div%2 == 0:
        return 'Not Prime'
    elif x == 3:
        return 'Prime'
    else:
        for i in range(3,div, 2):
            if div%i == 0:
                return 'Not Prime'
            else:
                return 'Prime'
```

```python
df1['Checking Primes'] = df1.apply(lambda x: ascii_function ( x['Title'], x['Votes']), axis=1)
display(df1)
```

```python
primes_df = df1[df1['Checking Primes']=='Prime']
primes_df
```

üêº üêº üêº üêº üêº Challenge 5. And finally some fantasy
Feel free to propose your own ranking based in aggregations of at least 3 columns of the dataset.

I'm going to create a ranking for Sci-Fi, Action or Fantasy genre films based on the votes received and the year of the film. The reason is that CGI technology improves with time so the newests films are advanced. 

If the year is less than 2009:
    - If votes are less than 5, no extra votes added. 
    - If votes are between 5 and 8, 2 extra votes added
    - If votes are greater than 8, 3 extra votes added. 
    
If the year is between 2009 and 2013:
    - If votes are less than 5, no extra votes added. 
    - If votes are between 5 and 8, 1 extra votes added
    - If votes are greater than 8, 2 extra votes added. 
    
If the year is greater than 2013:
    - If votes are less than 5, no extra votes added. 
    - If votes are between 5 and 8, no extra votes added
    - If votes are greater than 8, 1 extra votes added. 

```python
print(df1['Year'].min())
print(df1['Year'].max())
```

```python
def my_ranking(x,y,z):
    if 'Sci-Fi' not in x or 'Action' not in x or 'Fantasy' not in x:
        return z
    else:
        if y <= 2009:
            if z < 5:
                return z
            elif 5 <= z < 8:
                return z + 2
            elif z >= 8:
                return z + 3
        elif 2009 < y <=2013:
            if z < 5:
                return z
            elif 5 <= z < 8:
                return z + 1
            elif z >= 8:
                return z + 2
        elif y > 2013:
            if z < 8:
                return z
            elif z >= 8:
                return z + 1
```

```python
df1['Gon Ranking'] = df1.apply(lambda x: my_ranking ( x['Genre'], x['Year'], x['Votes']), axis=1)
display(df1)
```

üêº üêº üêº üêº üêº üêº Challenge 6. Freaky bonus
We want to know which movies might have hidden paterns in their description. A way to know that is finding those movies which the sum of all numeric values of the string description hash (SHA256) are between their revenue and their number of votes.

```python
def description_pattern(x,y,z):
    hs = hashlib.sha256(x.encode('utf-8')).hexdigest()
    suma = 0
    for i in hs:
        if i.isdigit():
            suma += int(i)
    
    if y < suma < z:
        return 'Pattern'
    else:
        return 'Not pattern'
    
```

```python
df1['Checking Pattern'] = df1.apply(lambda x: description_pattern ( x['Description'], x['Revenue (Millions)'], x['Votes']), axis=1)
display(df1)
```
